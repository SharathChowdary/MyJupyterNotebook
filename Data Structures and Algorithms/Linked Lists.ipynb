```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
class LinkedList:
    def __init__(self):
        self.head=None
    def add_at_beginning(self,data):
        if self.head==None:
            self.head=Node(data)
        else:
            node=Node(data)
            node.next=self.head
            self.head=node
    def insert_at_position(self,pos,data):
        length=self.get_length()
        if pos<0 or pos>length:
            print("Not possible to insert at that position")
            return 
        if pos==0:
            self.add_at_beginning(data)
        elif pos==length:
            self.add_at_end(data)
        else:
            count=0
            node=Node(data)
            temp=self.head
            while count<pos-1:
                temp=temp.next
                count+=1
            node.next=temp.next
            temp.next=node
    def add_at_end(self,data):
        if self.head==None:
            self.head=Node(data)
            return 
        temp=self.head
        while temp:
            if temp.next==None:
                temp.next=Node(data)
                break
            temp=temp.next
    def insert_after_value(self,data_after,data_to_insert):
        temp=self.head
        while temp:
            if temp.data==data_after:
                node=Node(data_to_insert)
                node.next=temp.next
                temp.next=node
                return
            temp=temp.next
        print(f"{data_after} not found")
    def insert_multiple_elements_at_beg(self,elements):
        if len(elements)==0:
            return
        head_node=self.head
        temp=Node(elements[0])
        starting_node=temp
        for element in elements[1:]:
            temp.next=Node(element)
            temp=temp.next
        temp.next=head_node
        self.head=starting_node
    def insert_multiple_elements_at_end(self,elements):
        for element in elements:
            self.add_at_end(element)
    def insert_multiple_elements_at_pos(self,pos,elements):
        if pos<0 or pos>self.get_length():
            print("Cannot insert at that position")
            return 
        if pos==0:
            self.insert_multiple_elements_at_beg(elements)
            return
        count=0
        temp=self.head
        while count<pos-1:
            count+=1
            temp=temp.next
        nxt=temp.next
        for element in elements:
            temp.next=Node(element)
            temp=temp.next
        temp.next=nxt
    def remove_at_beginning(self):
        if self.head==None:
            print("LinkedList is already empty")
            return
        self.head=self.head.next
    def remove_at_position(self,pos):
        length=self.get_length()
        if pos<0 or pos>=length:
            print(f"Please enter valid index between 0 and {length-1}")
            return 
       	if pos==0:
            self.remove_at_beginning()
        elif pos==length-1:
            self.add_at_end()
        else:
            count=0
            temp=self.head
            while count<pos-1:
                temp=temp.next
                count+=1
            temp.next=temp.next.next
    def remove_at_end(self):
        if self.head==None:
            print("LinkedList is already empty")
            return
        temp=self.head
        if temp.next==None:
            self.head=None
            return
        while(temp):
            if temp.next.next==None:
                temp.next=None
            temp=temp.next
    def remove_element_by_value(self,data):
        if self.get_index(data)==0:
            self.remove_at_beginning()
            return
        temp=self.head
        prev=None
        while temp:
            if temp.next.data==data:
                temp.next=temp.next.next
                return
            temp=temp.next
        print("Element not found")
    def print_all_nodes(self):
        if self.head==None:
            print("LinkedList is empty")
            return
        itr=self.head
        while(itr):
            print(itr.data,end="-->")
            itr=itr.next
        print()
    def get_index(self,data):
        count=0
        temp=self.head
        while temp:
            if temp.data==data:
                return count
            temp=temp.next
            count+=1
        return -1
    def get_length(self):
        temp=self.head
        length=0
        while temp:
            length+=1
            temp=temp.next
        return length
    def count(self,data):
        temp=self.head
        count=0
        while temp:
            if temp.data==data:
                count+=1
            temp=temp.next
        return count
    def __len__(self):
        return self.get_length()
    def element_at_index(self,index):
        if index<0 or index>=self.get_length():
            return None
        count=0
        temp=self.head
        while count<index:
            temp=temp.next
            count+=1
        return temp.data
    def is_element_there(self,value):
        temp=self.head
        while temp:
            if temp.data==value:
                return True
            temp=temp.next
        return False
lst=LinkedList()
lst.add_at_beginning(5)
lst.add_at_beginning(4)
lst.add_at_beginning(3)
lst.add_at_beginning(2)
lst.add_at_beginning(1)
lst.add_at_end(6)
lst.add_at_end(7)
lst.add_at_end(8)
lst.insert_at_position(3,"Three")
lst.remove_at_beginning()
lst.remove_at_end()
lst.remove_at_position(3)
lst.insert_multiple_elements_at_end([8,9,10])
lst.insert_multiple_elements_at_pos(0,["Four","Five"])
lst.print_all_nodes()
print("\nLength : ",lst.get_length())
lst.insert_after_value(2,"Two")
lst.print_all_nodes()
lst.remove_element_by_value(6)
lst.print_all_nodes()
```

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.prev=None
        self.next=None 
class DoublyLinkedList:
    def __init__(self):
        self.head=None
        self.tail=None
    def insert_at_beginning(self,data):
        if self.head==None:
            self.head=Node(data)
            self.tail=self.head
            return
        node=Node(data)
        node.next=self.head
        self.head.prev=node
        self.head=node
    def insert_at_end(self,data):
        if self.head==None:
            self.head=Node(data)
            return
        node=Node(data)
        node.prev=self.tail
        self.tail.next=node
        self.tail=node
    def insert_at_pos(self,pos,data):
        length=self.get_length()
        if pos<0 or pos>length:
            print("Cannot insert at that position")
            return
        if pos==0:
            self.insert_at_beginning(data)
        elif pos==length:
            self.insert_at_end(data)
        else:
            temp=self.head
            count=0
            while count<pos-1:
                temp=temp.next
                count+=1
            node=Node(data)
            node.next=temp.next
            temp.next.prev=node
            node.prev=temp
            temp.next=node
    def remove_at_beg(self):
        if self.head==None:
            print("LinkedList is already empty")
            return
        self.head=self.head.next
        self.head.prev=None
    def remove_at_end(self):
        if self.head==None:
            print("LinkedList is already empty")
            return
        self.tail=self.tail.prev
        self.tail.next=None
    def remove_at_pos(self,pos):
        length=self.get_length()
        if pos<0 or pos>=length:
            print("Please enter valid index")
            return
        if pos==0:
            self.remove_at_beg()
        elif pos==length-1:
            self.remove_at_end()
        else:
            temp=self.head
            count=0
            while count<pos-1:
                temp=temp.next
                count+=1
            temp.next.next.prev=temp
            temp.next=temp.next.next
    def insert_multiple_elements_at_beg(self,elements):
        temp=self.head
        node=Node(elements[0])
        starting_node=node
        for element in elements[1:]:
            new_node=Node(element)
            node.next=new_node
            new_node.prev=node
            node=node.next
        node.next=temp
        temp.prev=node
        self.head=starting_node
    def insert_multiple_elements_at_pos(self,pos,elements):
        length=self.get_length()
        if pos<0 or pos>length:
            print("Please enter valid index")
            return
        if pos==0:
            self.insert_multiple_elements_at_beg(elements)
        else:
            temp=self.head
            count=0
            while count<pos-1:
                temp=temp.next
                count+=1
            for element in elements:
                node=Node(element)
                node.next=temp.next
                if temp.next!=None:
                    temp.next.prev=node
                else:
                    self.tail=node
                temp.next=node
                node.prev=temp
                temp=temp.next
                
    def get_length(self):
        count=0
        temp=self.head
        while temp:
            temp=temp.next
            count+=1
        return count
    def forward_printing(self):
        temp=self.head
        while temp:
            print(temp.data,end="-->")
            temp=temp.next
        print()
    def backward_printing(self):
        temp=self.tail
        while(temp):
            print(temp.data,end="-->")
            temp=temp.prev
        print()
        print()
lst=DoublyLinkedList()
lst.insert_at_beginning(3)
lst.insert_at_beginning(2)
lst.insert_at_beginning(1)
lst.insert_at_end(4)
lst.insert_at_end(5)
lst.insert_at_end(6)
lst.insert_at_end(7)
lst.forward_printing()
lst.backward_printing()
lst.insert_at_pos(3,"Three")#adding element
lst.forward_printing()
lst.backward_printing()
lst.remove_at_beg()#removing at beg
lst.forward_printing()
lst.backward_printing()
lst.remove_at_end()#removing at end
lst.forward_printing()
lst.backward_printing()
lst.remove_at_pos(3)#removing at pos 3
lst.forward_printing()
lst.backward_printing()
lst.insert_multiple_elements_at_beg(["One","Two"])# inserting multiple elements at beg
lst.forward_printing()
lst.backward_printing()
lst.insert_multiple_elements_at_pos(3,["temp1","temp2"])# inserting multiple elements at pos 3
lst.forward_printing()
lst.backward_printing()
```
