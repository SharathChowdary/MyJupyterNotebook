# Dealing with NaN values

```python
dct={'cricket':[85,63,55,np.nan,78],'football':[96,np.nan,52,np.nan,69],'Basketball':[68,2,23,41,96]}
data=pd.DataFrame(data=dct)
data
```

```python
#dropping rows having NaN values
data.dropna()
```

```python
#dropping columns having NaN values
data.dropna(axis=1)
```

```python
#dropping rows having particular number of NaN values
data.dropna(thresh=2)
```

```python
#filling NaN with given value
data.fillna(value=333) #data.fillna(333)-->also valid and strings can also be passed
```

```python
#filling NaN values of cricket with mean of cricket
data['cricket'].fillna(value=data['cricket'].mean())
```

```python
#filling all NaN values with means of their columns
for i in data:
    data[i].fillna(value=data[i].mean(),inplace=True)
data
```

# Merging

merge() can only use columns (plus row-indices) and it is semantically suitable for database-style operations.

```python
table1=pd.DataFrame({'custID':['1001','1002','1003','1004'],'Q1':[101,102,103,104],'Q2':[201,202,203,204]})
table1
```

```python
table2=pd.DataFrame({'custID':['1001','1006','1003','1004'],'Q3':[301,302,303,304],'Q4':[401,402,403,404]})
table2
```

```python
#it merges only common custID in both table
pd.merge(table1,table2,on='custID')
```

```python
#it merges all columns and rows
pd.merge(table1,table2,how='outer',on='custID')
```

# Joining

pd.DataFrame.join() merely offers a shorthand for a subset of the use cases of .merge()

```python
table3=pd.DataFrame({'Q1':[101,102,103,104],'Q2':[201,202,203,204]},index=[1,2,3,4])
table3
```

```python
table4=pd.DataFrame({'Q3':[301,302,303,304],'Q4':[401,402,403,404]},index=[4,5,6,7])
table4
```

```python
#left join
table3.join(table4)
```

```python
#outer join
table3.join(table4,how='outer')
```

```python
#inner join
table3.join(table4,how='inner')
```
