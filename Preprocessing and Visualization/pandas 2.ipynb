# Merge() function

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
#merging two dataframes
df=pd.merge(df1,df2,on="city")
df
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
#by default it performs inner join
df=pd.merge(df1,df2,on="city")
df
```

```python
#outer join
df=pd.merge(df1,df2,on="city",how='outer')
df
```

```python
#left join
df=pd.merge(df1,df2,on="city",how='left')
df
```

```python
#right join
df=pd.merge(df1,df2,on="city",how='right')
df
```

```python
#indicator flag--> useful to indicate the row belongs to which dataframe
#outer join
df=pd.merge(df1,df2,on="city",how='outer',indicator=True)
df
```

```python
#Adding suffixes
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
#default suffixes are "_x" and "_y"
df=pd.merge(df1,df2,on='city',how="outer")
df
```

```python
#adding our own suffix
df=pd.merge(df1,df2,on='city',how="outer",suffixes=["_left","_right"])
df
```

# pivot_table() function
pivot table is used to summarize and aggregate the data inside the dataframe

```python
lst=[
("5/1/2017","new york",65,56),
("5/1/2017","new york",61,54),
("5/2/2017","new york",70,60),
("5/2/2017","new york",72,62),
("5/1/2017","mumbai",75,80),
("5/1/2017","mumbai",78,83),
("5/2/2017","mumbai",82,85),
("5/2/2017","mumbai",80,26)
]
data=pd.DataFrame(lst,columns=["date","city","temperature","humidity"])
data
#here the multiple values are there for a single date
```

```python
#mean values are returned for the same dates
#by default "aggfunc" is 'mean'
data.pivot_table(index='city',columns='date')
```

```python
#getting sum of values for the same day
data.pivot_table(index='city',columns='date',aggfunc='sum')
```

```python
#getting the count of values for the same date
data.pivot_table(index='city',columns='date',aggfunc='count')
```

```python
#getting aggregate values for all rows and columns
data.pivot_table(index='city',columns='date',margins=True)
```

# shift() function

```python
import pandas as pd
lst=[
("15-Aug-17",171),
("16-Aug-17",170),
("17-Aug-17",166.91),
("18-Aug-17",167.41),
("21-Aug-17",167.78),
("22-Aug-17",169.64),
("23-Aug-17",168.71),
("24-Aug-17",167.74),
("25-Aug-17",166.32),
("28-Aug-17",167.24)
]
data=pd.DataFrame(lst,columns=["Date","Price"])
data
```

```python
data.Date=pd.to_datetime(data.Date)
data=data.set_index('Date')
data
```

```python
#shifting prices to next day i.e. price of aug 15 is shifted to aug 16
data.shift(1)
```

```python
#shifting prices to 2 days i.e. price of aug 15 is shifted to aug 17
data.shift(2)
```

```python
##shifting prices to previous day i.e. price of aug 16 is shifted to aug 15
data.shift(-1)
```

```python
#creating a column having previous day prices
data['previous day price']=data.shift(1)
data
```

```python
#fetching the price change
data['price change']=data['Price']-data['previous day price']
data
```

```python
data['5 day % return']=(data['Price']-data['Price'].shift(-5))*100/data['Price'].shift(-5)
data
```
