# Seaborn

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.
It is built on top of matplotlib

```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
```

```python
#loading a dataset using seaborn
fmri=sns.load_dataset('fmri')
```

```python
#Displaying first 5 rows of the data
fmri.head()
```

```python
#Displaying last 5 contents of the data
fmri.tail()
```

```python
#Basic lineplot
sns.lineplot(x='timepoint',y='signal',data=fmri)
plt.show()
```

```python
#splitting lines based on event column
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event')
plt.show()
```

```python
#adding styles to the lines based on event column
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event',style='event')
plt.show()
```

```python
#adding markers
sns.lineplot(x='timepoint',y='signal',data=fmri,marker=True,hue='event')#marker or markers
plt.show()
```

# Barplots

```python
#loading pokemon dataset using "read_csv"
pokemon=pd.read_csv('pokemon.csv')
```

```python
pokemon.head()
```

```python
#Basic barplot
sns.barplot(x='Legendary',y='Speed',data=pokemon)
plt.title("Pokemon")
plt.show()
```

```python
#adding hue
sns.barplot(x='Legendary',y='Speed',data=pokemon,hue="Generation")
plt.title("Pokemon")
plt.show()
```

```python
#adding color
sns.barplot(x='Legendary',y='Speed',data=pokemon,color='green')
plt.show()
```

```python
#adding palettes
sns.barplot(x='Legendary',y='Speed',data=pokemon,palette='vlag')#vlag or Blues_d or rocket
plt.show()
```

# Scatter plot

```python
#loading iris dataset
iris=pd.read_csv('iris.csv')
```

```python
iris.head()
```

```python
#basic scatterplot
sns.scatterplot(x='sepal.length',y='petal.length',data=iris)
plt.show()
```

```python
#adding hue based on veriety
sns.scatterplot(x='sepal.length',y='petal.length',data=iris,hue='variety')
plt.show()
```

```python
#adding hue based on petal.length
sns.scatterplot(x='sepal.length',y='petal.length',data=iris,hue='petal.length')
plt.show()
```

```python
#adding style based on veriety
sns.scatterplot(x='sepal.length',y='petal.length',data=iris,hue='petal.length',style='variety')
plt.show()
```

# Histogram / Distribution plot

```python
#loading diamonds dataset
diamonds=pd.read_csv('diamonds.csv')
```

```python
diamonds.head()
```

```python
#Basic plot (both histogram and distribution plot)
sns.distplot(diamonds['price'])
plt.show()
```

```python
#to get only histogram
sns.distplot(diamonds['price'],kde='False')
plt.show()
```

```python
#to get only distribution plot
sns.distplot(diamonds['price'],hist='False')
plt.show()
```

```python
#setting a color
sns.distplot(diamonds['price'],kde='False',color='red')
plt.show()
```

```python
#adding bins
sns.distplot(diamonds['price'],kde='False',bins=10)
plt.show()
```

```python
#plotting vertically
sns.distplot(diamonds['price'],vertical='True')
plt.show()
```

# Joint plot

It is a combination of histogram and scatterplot

```python
#basic joint plot
sns.jointplot(x='sepal.length',y='petal.length',data=iris)
plt.show()
```

```python
#changing color
sns.jointplot(x='sepal.length',y='petal.length',data=iris,color='red')
plt.show()
```

```python
#Adding regression line using "kind" parameter
sns.jointplot(x='sepal.length',y='petal.length',data=iris,kind='reg')
plt.show()
```

# Box plots

```python
iris.head()
```

```python
#Basic boxplot
sns.boxplot(x='variety',y='sepal.length',data=iris)
plt.show()
```

```python
#setting a palette
sns.boxplot(x='variety',y='sepal.length',data=iris,palette='Set1')
plt.show()
```

```python
#changing width of borders using "linewidth" parameter
sns.boxplot(x='variety',y='sepal.length',data=iris,linewidth=3)
plt.show()
```

```python
#changing the order using "order" parameter
sns.boxplot(x='variety',y='sepal.length',data=iris,order=['Virginica','Setosa','Versicolor'])
plt.show()
```

```python
#another way of plotting graphs in seaborn
sns.boxplot(x=iris['variety'],y=iris['sepal.length'])
plt.show()
```

# Count plot

```python
sns.countplot(iris['variety'])
plt.show()
```
